```
import pandas as pd
import numpy as np
```

```
#Importando e vizualizando o arquivo csv
df_dcc = pd.read_csv('/work/Fifa-2021.csv')
df_dcc
```

```
#lista de todas posiçoes
list(df_dcc['team_position'].drop_duplicates())
```

```
#Pegando apenas os atacantes
posicoes = ['ST','CF']
```

```
#Aplicando o filtro atraves da lista 'posicoes' que criamos
selecao = df_dcc['team_position'].isin(posicoes)
df_dcc_atk = df_dcc[selecao]
df_dcc_atk
```

```
#Ajeitando o index
df_dcc_atk.index = range(df_dcc_atk.shape[0])
df_dcc_atk
```

```
#pegando apenas as colunas que vamos usar
df_dcc1 = df_dcc_atk[['sofifa_id','short_name','long_name','age','pace','shooting','dribbling','attacking_crossing','attacking_finishing','attacking_heading_accuracy','attacking_short_passing','attacking_volleys','skill_ball_control','movement_sprint_speed','movement_agility','movement_reactions','movement_balance','power_shot_power','power_long_shots','mentality_vision']]
df_dcc1
```

```
#vizualizando algumas metricas
df_dcc1.describe()
```

```
#sumarizando as colunas com os atributos dos jogadores
sum_colunas = df_dcc1['pace']+ df_dcc1['shooting'] + df_dcc1['dribbling'] + df_dcc1['attacking_crossing'] + df_dcc1['attacking_finishing'] + df_dcc1['attacking_heading_accuracy'] + df_dcc1['attacking_short_passing'] + df_dcc1['attacking_volleys'] + df_dcc1['skill_ball_control'] + df_dcc1['movement_sprint_speed'] + df_dcc1['movement_agility'] + df_dcc1['movement_reactions'] + df_dcc1['movement_balance'] + df_dcc1['power_shot_power'] + df_dcc1['power_long_shots'] + df_dcc1['mentality_vision']
```

```
# criando uma nova coluna atraves da sumarizacao que fizemos acima
df_dcc1['pt_total'] = sum_colunas/16
df_dcc1
```

```
df_dcc1['craques'] = np.where(df_dcc1['pt_total']>=70, '1', '0')
df_dcc1
```

```
#Dropando o sofifa_id para refazer o id dos jogadores
df_dcc1.drop(['sofifa_id'], axis=1)
```

```
#Colocando a coluna 'sofifa_id' novamente, porem com os id dos jogadores refeitos
df_dcc1['sofifa_id'] = range(df_dcc1.shape[0])
```

```
#Tirando a coluna de 'short_name'
df_dcc1.drop(['short_name'], axis=1, inplace=True)
```

```
#Criando um novo dataframe para adicionar os novos jogadores
novos_jogadores = pd.DataFrame(columns = [['age', 'pace','shooting', 'dribbling','attacking_crossing', 'attacking_finishing','attacking_heading_accuracy', 'attacking_short_passing', 'attacking_volleys','skill_ball_control','movement_sprint_speed', 'movement_agility', 'movement_reactions', 'movement_balance', 'power_shot_power','power_long_shots','mentality_vision','pt_total']])
novos_jogadores
```

```
#Ajeitando o index
df_dcc1.index = range(df_dcc1.shape[0])
df_dcc1
```

```
#Separando o dataframe
df_dcc1.to_csv('df_dcc1.csv', index=False)
```

```
#Separando o dataframe para adicionar novos jogadores
novos_jogadores.to_csv('df_novos_jogadores', index=False)
```
