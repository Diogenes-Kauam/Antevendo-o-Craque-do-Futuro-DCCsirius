```
#importando bibliotecas para manipular os data_Frames
import pandas as pd
import numpy as np

#importando bibliotecas para criacao e aplicacao da inteligencia artificial
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

```
#importado e vizualizando o data_frame
df_dcc = pd.read_csv('/work/df_dcc1.csv')
df_novos_jogadores = pd.read_csv('/work/df_novos_jogadores')
df_dcc
```

```
df_novos_jogadores
```

```
#Pegando a coluna que queremos prever
y = df_dcc['craques']

#Pegando apenas as colunas que vamos usar nessa previsao
x = df_dcc.drop(columns=['craques', 'long_name','sofifa_id'], axis=1)

#Dividindo o data_frame em 4 partes: x_treino, x_teste, y_treino, y_teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size= 0.7, random_state=1)
```

```
#Criando a inteligencia artificial
modelo_regressaolinear = LinearRegression()
modelo_arvorededecisao = RandomForestRegressor(random_state=1)

#Treinando a inteligencia artificial
modelo_arvorededecisao.fit(x_treino, y_treino)
modelo_regressaolinear.fit(x_treino, y_treino)
```

```
#Importando o 'metrics' para fazer as previsoes
from sklearn import metrics

# Criando as previsoes
previsao_regressao = modelo_regressaolinear.predict(x_teste)
previsao_arvore = modelo_arvorededecisao.predict(x_teste)

#Comparar os modelos
print('{:.2f}'.format((metrics.r2_score(y_teste, previsao_regressao))))
print('{:.2f}'.format((metrics.r2_score(y_teste, previsao_arvore))))
```

```
lista_jogador = []
def criar_jogador():
    
    idade = int(input('Digite a idade do jogador:'))
    while idade > 50 or idade < 12:
        idade = int(input('Jogador muito velho/muito novo. Digite a idade novamente:'))
    lista_jogador.append(idade)
    
    pace = int(input('Digite o nivel de ritmo do jogador (1 a 100):'))
    while pace > 100 or pace < 0:
        pace = int(input('Nivel fora do permitido. Digite novamente:'))
    lista_jogador.append(pace)
    
    shooting = int(input('Digite o nivel de arrancada do jogador (1 a 100):'))
    while shooting > 100 or shooting < 0:
        shooting = int(input('Nivel fora do permitido. Digite novamente:'))
    lista_jogador.append(shooting)

    dribbling = int(input('Digite o nivel de drible do jogador (1 a 100):'))
    while dribbling > 100 or dribbling < 0:
        dribbling = int(input('Nivel fora do permitido. Digite novamente:'))
    lista_jogador.append(dribbling)

    attacking_crossing = int(input('Digite o nivel do cruzamento do jogador (1 a 100):'))
    while attacking_crossing > 100 or attacking_crossing < 0:
        attacking_crossing = int(input('Nivel fora do permitido. Digite novamente:'))
    lista_jogador.append(attacking_crossing)

    attacking_finishing = int(input('Digite o nivel de finalização do jogador (1 a 100):'))
    while attacking_finishing > 100 or attacking_finishing < 0:
        attacking_finishing = int(input('Nivel fora do permitido. Digite novamente:'))
    lista_jogador.append(attacking_finishing)

    attacking_heading_accuracy = int(input('Digite o nivel de direcionamento ao ataque do jogador (1 a 100):'))
    while attacking_heading_accuracy > 100 or attacking_heading_accuracy < 0:
        attacking_heading_accuracy = int(input('Nivel fora do permitido. Digite novamente:'))
    lista_jogador.append(attacking_heading_accuracy)

    attacking_short_passing = int(input('Digite o nivel de direcionamento ao ataque do jogador (1 a 100):'))
    while attacking_short_passing > 100 or attacking_short_passing < 0:
        attacking_short_passing = int(input('Nivel fora do permitido. Digite novamente:'))
    lista_jogador.append(attacking_short_passing)

    attacking_volleys = int(input('Digite o nivel de voleio do jogador (1 a 100):'))
    while attacking_volleys > 100 or attacking_volleys < 0:
        attacking_volleys = int(input('Nivel fora do permitido. Digite novamente:'))
    lista_jogador.append(attacking_volleys)

    skill_ball_control = int(input('Digite o nivel de controle de bola do jogador (1 a 100):'))
    while skill_ball_control > 100 or skill_ball_control < 0:
        skill_ball_control = int(input('Nivel fora do permitido. Digite novamente:'))
    lista_jogador.append(skill_ball_control)

    movement_sprint_speed = int(input('Digite o nivel de velocidade do jogador (1 a 100):'))
    while movement_sprint_speed > 100 or movement_sprint_speed < 0:
        movement_sprint_speed = int(input('Nivel fora do permitido. Digite novamente:'))
    lista_jogador.append(movement_sprint_speed)

    movement_agility = int(input('Digite o nivel de agilidade do jogador (1 a 100):'))
    while movement_agility > 100 or movement_agility < 0:
        movement_agility = int(input('Nivel fora do permitido. Digite novamente:'))
    lista_jogador.append(movement_agility)

    movement_reactions = int(input('Digite o nivel de reação de movimento do jogador (1 a 100):'))
    while movement_reactions > 100 or movement_reactions < 0:
        movement_reactions = int(input('Nivel fora do permitido. Digite novamente:'))
    lista_jogador.append(movement_reactions)

    movement_balance = int(input('Digite o nivel de equilibrio do jogador (1 a 100):'))
    while movement_balance > 100 or movement_balance < 0:
        movement_balance = int(input('Nivel fora do permitido. Digite novamente:'))
    lista_jogador.append(movement_balance)

    power_shot_power = int(input('Digite o nivel de poder de tiro do jogador (1 a 100):'))
    while power_shot_power > 100 or power_shot_power < 0:
        power_shot_power = int(input('Nivel fora do permitido. Digite novamente:'))
    lista_jogador.append(power_shot_power)

    power_long_shots = int(input('Digite o nivel de poder de tiros longos do jogador (1 a 100):'))
    while power_long_shots > 100 or power_long_shots < 0:
        power_long_shots = int(input('Nivel fora do permitido. Digite novamente:'))
    lista_jogador.append(power_long_shots)

    mentality_vision = int(input('Digite o nivel de mentalidade do jogador (1 a 100):'))
    while mentality_vision > 100 or mentality_vision < 0:
        mentality_vision = int(input('Nivel fora do permitido. Digite novamente:'))
    lista_jogador.append(mentality_vision)

    sum_colunas = pace + shooting + dribbling + attacking_crossing + attacking_finishing + attacking_heading_accuracy + attacking_short_passing + attacking_volleys + skill_ball_control + movement_sprint_speed + movement_agility + movement_reactions + movement_balance + power_shot_power + power_long_shots + mentality_vision
    lista_jogador.append(sum_colunas/16)

    df_novos_jogadores.loc[df_novos_jogadores.shape[0]] = lista_jogador
    print('Jogador criado com sucesso!')
```
